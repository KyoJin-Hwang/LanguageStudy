```python
# 두개의 정수를 입력 받아서 더하는 코드를 작성하시오.
# (단, 두개의 정수가 0이 들어올 때 까지 반복한다.)
num1 = num2 = -1

while num1 != 0 or num2 !=0 :
    num1 = int(input('첫번째 정수 입력>> '))
    num2 = int(input('첫번째 정수 입력>> '))
    
    if num1 != 0 or num2 !=0 :
        print(f'두 수의 합 : {num1+num2}')
print('프로그램 종료')
```
