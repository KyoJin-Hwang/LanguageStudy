```python
#list 슬라이싱 -> 리스트명[star인덱스:end인덱스]
list3=[1,8,7,3,5]
print(list3[1:3])
print(list3[:2])
print(list3[3:])
print(list3[3:4])
print(type(list3[3:4]))
print(list3[3])
print(type(list3[3]))
```

```python
list4 = [1,2,3]
list5 = [3,4,5,6]
print(list4+list5)
```

```python
list6 = [1,3,4,7,2,8]
list6[1] = 5
print(list6)
list6[1:4] = [6]
print(list6)
```

```python
list7 = [1,2,3,4,5]

# 리시트 array 숫자 5를 7로 바꾸기 
list7[-1] = 7
print(list7)

# 숫자2를 리스트 ['a','b','c']로 바꾸기
list7[1] = ['a','b','c']
print(list7)

# 숫자 4,7 를 'd','e','f','g'로 바꾸기 
list7[3:] = ['d','e','f','g']
print(list7)

#숫자 1을 'h','i','j'
list7[0:1] = ['h','i','j']
print(list7)
```

```python
#리스트 값 삽입 insert(인덱스값)
```

```python
s = "Hi, My name is Mihee"
inputData = input('검색할 문자를 입력하세요 : ')

if inputData in s:
    print(f"'{inputData}'는 {s.count(inputData)}번 들어있습니다. ")
else :
    print(f"'{inputData}'는 문자열에 들어있지않습니다.")
```

```python
lst = ['딸기','바나나', '수박', '체리', '포도']

inputdata = input('검색할 문자를 입력하세요 >> ')

if inputdata in lst:
    print(f'{inputdata}는 리스트에 {lst.index(inputdata)}번째 인덱스에 들어있습니다.')
else :
    print(f'{inputdata}는 리스티에 들어있지 않습니다.')
```
