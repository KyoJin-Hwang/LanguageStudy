```python
#       .
# (,   0    .)
num1 = num2 = -1

while num1 != 0 or num2 !=0 :
    num1 = int(input('  >> '))
    num2 = int(input('  >> '))
    
    if num1 != 0 or num2 !=0 :
        print(f'   : {num1+num2}')
print(' ')
```
