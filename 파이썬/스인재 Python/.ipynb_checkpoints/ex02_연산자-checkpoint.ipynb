```python
num3 = int(input("정수를 입력하세요 >> "))
num4 = int(input("정수를 입력하세요 >> "))
print()
print("더하기 결과 : " ,num3+num4)
print("뺴기 결과 : " ,num3-num4)
print("나누기 결과 : " ,num3/num4)
print("곱하기 결과 : " ,num3*num4)
print()
print(f"더하기 결과 : {num3+num4}")
print(f"뺴기 결과 : {num3-num4}")
print(f"나누기 결과 : {num3*num4}")
print(f"곱하기 결과 : {num3/num4}")
print()
print("더하기 결과 : {}" .format(num3+num4))
print("뺴기 결과 : {}" .format(num3-num4))
print("나누기 결과 : {}" .format(num3/num4))
print("곱하기 결과 : {}" .format(num3*num4))
print()
print("더하기 결과 : " +str(num3+num4))
print("뺴기 결과 : " +str(num3-num4))
print("나누기 결과 : " +str(num3/num4))
print("곱하기 결과 : " +str(num3*num4))
```

```python
time = int(input("시간 입력 >> "))
hour = int(time/3600)
minute = int(time%3600/60)
second = int(time%60)
print("{}시간 {}분 {}초".format(hour, minute, second))
```

```python
# 2번째 방법
time1 = int(input("시간 입력 >> "))
hour1 = time1//3600
minute1 = time1%3600//60
second1 = time1%60
print("{}시간 {}분 {}초".format(hour1,minute1,second1))
```
