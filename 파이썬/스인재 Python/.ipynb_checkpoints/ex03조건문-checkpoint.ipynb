```python
# 변수 num1과 num2에 숫자를 입력받고 크기를 비교하세요
num1 = int(input("첫번째 정수입력 >> "))
num2 = int(input("두번째 정수입력 >> "))

if num1>num2 :
    print("첫 번째 정수가 더 큽니다.")
elif num1<num2 :
    print("두 번째 정수가 더 큽니다.")
else :
    print("두 수가 똑같습니다.")
```

```python
# 변수 score에 점수를 입력 받아서 다음과 같이 학점을 부여하시오.
# 100이하,90이상 -> A
# 90미만, 80이상 -> B
# 80미만, 70이상 -> C
# 70미만, 60이상 -> D
# 60미만 -> F
score = int(input('점수 입력 >> '))

if score<=100 and score>=90 :
    print(f"{score}점은 A학점 입니다.")
elif score< 90 and score>=80 :
    print(f"{score}점은 B학점 입니다.")
elif score< 80 and score>=70 :
    print(f"{score}점은 C학점 입니다.")
elif score< 70 and score>=60 :
    print(f"{score}점은 D학점 입니다.")
elif score<60 :
    print(f"{score}점은 F학점 입니다.")
```

```python
#자판기 프로그램
#메뉴 출력
#돈 입력
#메뉴 선택
#결과 확인
#  - 잔액부족 출력
#    - 잔액 환전(1000원,500원,100원)
print('''
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++     SMHRD    +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')

print("음료 자판기 입니다. 먹고 싶은 음료를 고르세요(10000원권, 50000원권 지폐 사용불가)")
coin = int(input("insert coin >> "))
#1콜라
menu1 = 600
#2사이다
menu2 = 800
#3물
menu3 = 1000 

if coin>500 and coin<10000 :   
    menu = int(input("메뉴선택 >> "))
    if menu==1 :
        coin=coin-menu1
        money = coin//1000
        money1 = coin%1000//500
        money2 = coin%500//100 
        print(f"1000원 {money}개, 500원 {money1}개, 100원 {money2}개")
    elif menu==2 :
        coin=coin-menu2
        money = coin//1000
        money1 = coin%1000//500
        money2 = coin%500//100
        print(f"1000원 {money}개, 500원 {money1}개, 100원 {money2}개")
    elif menu==3 :
        coin=coin-menu3
        money = coin//1000
        money1 = coin%1000//500
        money2 = coin%500//100  
        print(f"1000원 {money}개, 500원 {money1}개, 100원 {money2}개")
elif coin<=500 :
        print("금액이 부족합니다. 다시 확인해주세요.")
elif coin>=10000 :
        print("잔액이 부족합니다. 투입하신 금액을 받아가세요.")
```
