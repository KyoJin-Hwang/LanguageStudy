### 인덱싱과 슬라이싱

```python
# 문자열 인덱싱

s4 = 'My name is MH'
print(s4[0])
print(s4[8])
print(s4[12])
print(s4[-1])
```

```python
print(s4[3]+s4[4]+s4[5]+s4[6])
```

```python
# 문자열 슬라이싱
s5 = 'My name is MH'

print(s5[0:4])
print(s5[3:11])
print(s5[3:15])
print(s5[-5:-1])
print()
# 0부터 잘라내고 싶을 때
print(s5[:4])
print(s5[3:])
print(s5[-5:])
print(s5[10:3])
```

```python
day = "2022년 3월 3일의 날씨는 맑음입니다."
print("날짜 : " + day[:11])
print("날씨 : " + day[-6:-4])
```

```python
month = 3
day = 23
s6 = "오늘은 "+str(month)+"월"+ str(day) +"일 입니다."
print(s6)
```

```python
# %기호 포매팅 -> %d를 활용해서 정수 대입

month = 3
day = 23
s6 = "오늘은 %d월 %d일 80%%입니다."%(month, day)
print(s6)
```

```python
# format함수 포매팅
month = 3
day = 23
s6 = "오늘은 {}월 {}일 입니다." .format(month, day)
print(s6)
```

```python
# f문자열 포매팅
month = 3
day = 23
s6 = f"오늘은 {month}월 {day}일 입니다."
print(s6)
```

```python
x=100
y=200
sum2=x+y
print(("{}와 {}의 합은 {}입니다.") .format(x,y,sum2))
print(("%d와 %d의 합은 %d입니다.") %(x,y,sum2))
print(f"{x}와 {y}의 합은 {sum2}입니다.")
```

### 문자열 관련 함수


```python
s7 = "My name is Mihee"

# 문자열에 포함된 문자 개수 세기
print(s7.count('e'))
print(s7.count("Mihee"))

# 문자 위치 알려주기
print(s7.index("e"))
print(s7.find("e"))

# 대소문자 변환
print(s7.upper())
print(s7.lower())

# 문자열 바꾸기
print(s7.replace("Mihee","KimMihee"))
print(s7.rfind("e"))

# 문자열 자르기(리스트타입으로 반환된다.)
print(s7.split())
```
