```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. 영화리뷰사이트 요청 req활용해서 
res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=164122&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
```

```python
# 2. 응답코드 확인 후 , bs으로 데이터를 변형
soup = bs(res.text, "lxml")
```

```python
# 3. select기능으로 활용해서 영화 리뷰텍스트 정보만 수집
# 3-1 참고! 원하는 요소의 위치(태그정보)확인하는 방법 
#  우클릭검사
# F12 > 셀렉트버튼 > 요소 클릭
review = soup.select("div.score_reple > p > span ")
```

```python
# 리뷰데이터에 접근이 불가능한 이유
# iframe을 통해서 다른서버에 있는 정보를 빌려다가 보여주기만 할뿐!
# 실제 데이터가 저장이 되어있는 서버의 주소로 찾아가기!
# F12 > iframe 검색 >  실제 사용되는 iframe 찾기
```

```python
# 1. 선택자를 정확하게 쓰는 방법이 제일 best 
# 2. p > span:nth-child(2)  < nth-child 가상선택자 > 
# 3. 사용할수 없는 이유 : 관람객 글자가 없는 경우가 존재(span 하나만존재)
# 불피요한 관람객 글자 추출해서 제거 * 많이 사용안됨
```

```python
viewer = soup.select("span.ico_viewer")
```

```python
for i in viewer :
    i.extract()
```

```python
# 불필요한 개행문자가 있는경우에는 제거 .strip()
for i in review:
    print(i.text.strip())
```

### 1페이지부터 10페이지까지 수집

```python
# 반복문 진행상황 로딩바형태로 체크
from tqdm import tqdm_notebook as tq
```

```python
for i in tq(range(1,11)):
    res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=164122&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(i))
    soup = bs(res.text, "lxml")
    viewer = soup.select("span.ico_viewer")
    for i in viewer:
        i.extract()
    review = soup.select("div.score_reple > p")
    for i in review :
        print(i.text.strip())
```
